```python
import matplotlib.pyplot as plt
import numpy as np

file = open("overhead_output/metasimclr.txt", "r")
lines = file.readlines()
cpus = []
mems = []
times = []
prev_time = -1
for line in lines:
    if "python" in line:
        stats = line.split()
        cpu = float(stats[8])
        mem = float(stats[9])/100*12*1024
        time_min = float(stats[10].split(":")[0])
        time_sec = float(stats[10].split(":")[1])
        time = time_min*60 + time_sec
        if time < prev_time:
            continue
        prev_time = time
        cpus.append(cpu)
        mems.append(mem)
        times.append(time)

file.close()

plt.figure()
plt.plot(cpus, label="CPU")
plt.plot(mems, label="Memory")
plt.plot(times, label="Time")
plt.xlabel("Number of epochs")
plt.ylabel("Value")
plt.title("MetaSimCLR overhead")
plt.legend()
plt.show()

# from 5 to 10 idle
# from 23 to 37 adaptation
# from 48 to end fine tune

idle_mems = mems[16:18]
idle_cpus = cpus[16:18]
idle_mem = np.mean(idle_mems)
idle_cpu = np.mean(idle_cpus)

adapt_mems = mems[23:37]
adapt_cpus = cpus[23:37]
print("Adaptation memory: ", np.mean(adapt_mems) - idle_mem)
print("Adaptation CPU: ", np.mean(adapt_cpus) - idle_cpu)

tune_mems = mems[48:]
tune_cpus = cpus[48:]
print("Fine tune memory: ", np.mean(tune_mems) - idle_mem)
print("Fine tune CPU: ", np.mean(tune_cpus) - idle_cpu)
```
